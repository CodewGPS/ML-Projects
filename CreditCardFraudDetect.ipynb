```python
#importing all the dependencies
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#creating a pandas dataframe
df=pd.read_csv('creditcard.csv')
```

```python
#first 5 rows of df
df.head()
```

```python
df.describe().mean()
```

```python
#data processing and cleaning
df.isnull().sum()
```

This shows that there are no null values in the df

```python
df['Class'].value_counts()
```

Unreasonably large dataset, will need to shorten the dataset.
1--> Fraud Transactions
0--> Legit Transactions

```python
#dividing data into legit transactions and fraud transactions
legit_tr=df[df.Class==0]
fraud_tr=df[df.Class==1]
```

```python
legit_tr.head()
```

```python
fraud_tr.head()
```

```python
legit_tr.Amount.describe()
```

```python
fraud_tr.Amount.describe()
```

Now we perform undersampling, that is taking a sample from the already avaialable data which can help in training our model more efficiently.

```python
#Undersampling
legit_sample=legit_tr.sample(n=500)
```

```python
legit_sample.head()
```

```python
new_df=pd.concat([legit_sample,fraud_tr])
```

```python
new_df.head()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
X=new_df.drop(columns='Class',axis=1)
Y=new_df['Class']
```

```python
X
```

```python
Y
```

Now we use scikit lab's built in module to split our data in train test. Stratify is used to order data accordingly in and ordered manner.

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

Now we train the model

```python
#Training the model
model=LogisticRegression(max_iter=1000)
```

```python
#fitting the data
model.fit(X_train,Y_train)
```

```python
X_train_predict=model.predict(X_train)
X_train_score=accuracy_score(X_train_predict,Y_train)
```

```python
#checking score
X_train_score
```

```python
#evaluating test data
X_test_predict=model.predict(X_test)
X_test_score=accuracy_score(X_test_predict,Y_test)
```
